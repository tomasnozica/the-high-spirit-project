## Word Cloud

Now, lets create some word clouds to visualize the data

```python
filename = "corpus.pkl"
data = []
outfile = open(filename,'rb')
corpus = pickle.load(outfile)
```

```python
text = ' '.join(corpus)
```

```python
# Terminal / Anaconda Prompt: conda install -c conda-forge wordcloud
from wordcloud import WordCloud

wc = WordCloud(background_color="white", colormap="Dark2",max_font_size=150, random_state=42)
wc.generate(text)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Funny how the most common word in motivational quotes is "don't" right? 

So, okay, if we think that one of the most motivational songs in history says "Don't Stop Believing" we can conclude that this words is not an outsider.

What is clear is that "make", "life","dream", "work","change" are inspiring words. 

SCENARIO 2: What happens with our Word Cloud if we delete the "stop words"?

```python
import pickle
filename = "data_clean.pkl"
data_clean = []
outfile = open(filename,'rb')
data_clean = pickle.load(outfile)
```

```python
text_clean = ' '.join(data_clean)
```

```python
from wordcloud import WordCloud

wc_clean = WordCloud(background_color="white", colormap="Dark2",max_font_size=150, random_state=42)
wc_clean.generate(text_clean)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(wc_clean, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Well, well... now this cloud word represents what our quotes are about! It's always very important to remove stopwords because they don't show in a meaningful way what our text is about.
